&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Преобразования, которые мы сейчас называем преобразованиями Радона, возникли впервые в работах Германа Минковского (Hermann Minkowski), Пола Функа (Paul Funk) и наконец в работах Иоганна Радона в 1904-ом, 1913-ом и 1917-ом годах соответственно. Минковский занимался изучением геомтрических объектов называемых *телами постоянной ширины* (теорема Минковского о телах постоянной ширины), Функ рассматривал преобразование функций на сфере $S^2$ (функция на $S^2\mapsto$ её интегралы по большим окружностям), а сам Радон рассмотрел эти преобразования на плоскости $R^2$, получил формулы обращения и обобщил эти формулы на произвольные размерности $n\geq 2$. Напомним, что классическое преобразование Радона $R$ определяется по следующей формуле:
\begin{equation}\label{radon-def}
    Rf(s,\theta) = \int\limits_{-\infty}^{+\infty} f(s\theta+ t\theta^\perp) \, dt, 
    \, s\in \mathbb{R}, \, \theta\in S^1.
\end{equation}

(Формальное-) Дуальное преобразование Радона определяется по следующей формуле:

\begin{equation}\label{eq:radon-adjoint-def}
    R^*g(x) = \int\limits_{S^1}g(x\theta, \theta) \, d\theta, \, x\in \mathbb{R}^2.
\end{equation}

###  1. Преобразование Радона на компьютере

#### 1.1 Дискретизация

Лишь для небольшого класса функций возможно аналитическое вычисление преобразования Радона, поэтому 
алгоритм вычисления $R$ для произвольных функций черезвычайно важен. 

Главный вопрос, как численно вычислять интеграл в формуле для $R$? 
По очевидным причинам при вычислениях все объекты (функции, преобразования и т. д.) представляются в памяти компьютера в виде конечных массивов. Поэтому для численной работы с преобразованиями Радона необходимо зафиксировать некоторую конечную дискретизацию с помощью которой мы будем представлять наши данные. 

Очевидно, что в нашем случае дискретизации должно быть две: 
1. дискретизация на цилиндре $R\times S^1$, т. е. представление функций $g(s,\theta)$ массивом значений
   $\{g_{ij}\}$, $i = \overline{1,N_s}$, $j=\overline{1, N_\theta}$ в точках 
   $\{(s_i, \theta_j) \in R\times S^1\}$
2. дискретизация на $R^2$, т. е. представление функций $f(x), \, x\in R^2$ множеством значений
   $\{f_i\}$, $i=\overline{1, N_p}$ (если $f$ - изображение на $N\times N$ пикселей, то $N_p = N^2$, соотвтетственно)

Конкретный выбор дискретизаций вообще говоря зависит от того, какие математические задачи нам необходимо решать: интерополяция, интегрирование, дифференцирование, использование спектральных методов (преобразование Фурье. Существует множество схем дискретизаций, оптимальных для каждой задачи, например 
   
   * нули полиномов Чебышова $T_k$ на интервале $[-a, a]$ - для 
     [интерполяции](https://en.wikipedia.org/wiki/Chebyshev_nodes) 
     (см., например, [эффект Рунге](https://en.wikipedia.org/wiki/Runge%27s_phenomenon))
   * [квадратуры Гаусса на интервале](https://en.wikipedia.org/wiki/Gaussian_quadrature) $[-a,a]$ - 
     для точного интегрирования
   * равномерные сетки в $R^n$ - для [спектральных методов](https://math.mit.edu/~stevenj/fft-deriv.pdf),
     эффективное использование FFT, iFFT
   

В данном задании нам будет достаточно простейших равномерных дискретизаций:

   * Во-первых, мы будем предполагать, что все функции $f(x)$ имеют носитель в круге $\{\|x\|_2 \leq 1\}$.
     Такое предположение мы используем, чтобы вместо цилиндра $R\times S^1$ рассмотреть ограниченное 
     множество $[-1,1]\times S^1$. Окружность $S^1$ представим в виде интервала $[0,2\pi)$ с 
     параметризацией $\theta = (\cos\varphi, \sin\varphi)$, $\varphi \in [0, 2\pi)$. На $[-1,1]$ будем 
     использовать равномерную сетку. 
   
     Для удобства представления функций возьмем квадрат $[-1,1]\times [-1,1]$ и будем также использовать 
     равномерную сетку.
     
     В итоге дискретизация на $[-1,1]\times [0, 2\pi)$, $[-1,1]\times[-1,1]$ выглядит так:
     
     \begin{align}
        & s_j = -R + (j-1)\Delta s, \, \varphi_i = \dfrac{2\pi (i-1)}{N_\theta}, \, j = \overline{1, N_s}, 
        \, i = \overline{1, N_\theta}, \\
        & x_i = -R + \Delta / 2 + (i-1) \Delta, \, y_i = -1.0 + \Delta / 2 + (i-1)\Delta, \, 
        \Delta = 2R / N, \, R = 1.0.
     \end{align}
     
     Смещение на $\Delta / 2$ связано с тем, что мы указываем центры координат соответствующих пикселей.


#### 1.2 Интегрирование вдоль луча на плоскости

Теперь нам необходимо определить, как представлять луч во введенных дискретизациях и интегрировать функции 
вдоль лучей. 

В параграфе 3 мы ещё увидим, как это делать "точно", а пока мы будем 
вычислять интегралы $Rf$ лишь приближенно. Выигрышь состоит в том, что алгоритм получается 
заметно проще, а ошибка интегрирования меняется незначительно. Проигрышь же состоит в том, что 
алгоритм будет работать достаточно медленно. В домашнем задании вам предлагается реализовать "точный" и 
одновременно "быстрый" алгоритм интегрирования самостоятельно.

Пусть мы хотим вычислить интеграл $f$ вдоль луча $(s,\theta(\varphi))$, $s\in [-1,1], \varphi \in [0,2\pi)$. 

Интегрирование вдоль луча это одномерное интегрирование по параметру $t, t\in [-\sqrt{2-s^2}, \sqrt{2-s^2}]$ (такой диапазон взят так как мы интегрируем вдоль всего луча, а не только внутри круга $\{\|x\|_2 \leq 1\}$), которое можно записать следующим образом:

\begin{align}
    &Rf(s,\varphi) = \int\limits_{-\sqrt{2-s^2}}^{\sqrt{2-s^2}}f(s\theta + t\theta^\perp)\, dt 
    \approx \Delta t\sum_{k=0}^{N(s,\varphi)} f(s\theta + t_k \theta^\perp), \theta = (\cos\varphi, \sin\varphi),
    \theta^\perp = (-\sin\varphi, \cos\varphi),\\
    &t_k = t_0 + k\Delta t, k = \overline{0, N(s,\varphi)-1}. 
\end{align}

Здесь мы выбрали равномерную сетку по $t$, и простое правило интегрирования порядка $0$, где функция представляется локально-постоянной вдоль луча. 

Чтобы вычислить вышеупомянутую сумму, нам надо ответить на два следующих вопроса:
  1. Какими брать $t_0, N(s,\varphi), \Delta t$ ?
  2. Как вычислять $f(s\theta + t_k \theta^\perp)$ ?

  1. Из рисунка выше, видно, что $t_0 = -\sqrt{2-s^2}$,

     или $\Delta t = \Delta / 2$. 
     Выбор шага между $\Delta$ или $\Delta/2$ имеет эмпирический характер

  2. Вычислять $f(s\theta + t_k \theta^\perp)$ очень просто - вычисляем координату

### 2. Двойственное преобразование Радона

Двойственное преобразование Радона можно вычислять путём дискретизации формулы для $R^*$:

$$
 R^*g (x) = \int\limits_{S^1}g(x\theta, \theta)\, d\theta \approx \Delta \varphi
 \sum_{i=1}^{N_\varphi} g(x\cdot \theta_i), \theta_i), \, \theta_i = \left(\cos\left((i-1)\frac{2\pi}{N_\varphi}\right), \sin\left((i-1)\frac{2\pi}{N_\varphi}\right)\right)
$$

Преобразование Радона $g(s,\theta) = Rf$ нам известно лишь на множестве $\{s_i\}_{i=1}^{N_s}\times \{\theta_i\}_{i=1}^{N_\theta}$, поэтому в формуле выше не все слагаемые нам известны (проблема в переменной $s$). Однако в данном случае, их можно получить при помощи одномерной (линейной) интерполяции на множестве $\{\theta_i\}\times [-1,1]$

Из форулы видно, что интерполяцию необходимо делать по одномерной переменной $s$, данные для $\theta_i$ нам известны. Так как мы будем вычислять $R^*g$ на диксретной сетке $\{x_j\}_{j=1}^{N^2}$ на $[-1,1]^2$, то для каждого $\theta_i$ наиболее эффективно делать интерполяцию сразу на всём массиве точек $\{x_j \cdot \theta_i\}_{j=1}^{N^2}$.

Для одномерной интерполяции можно использовать функцию ``numpy.interp(interp_points, points, values, left, right)``, где ``interp_points`` - координаты точек для интерполяции, ``points`` - координаты известных точек, ``values`` - значения функции в известных точках, ``left, right`` - граничные значения, если координаты интерполяции выходят за границы известных данных.

#### Замечание об интерполяции

Формально, можно задаться вопросом - какой метод интерполяции выбрать? Помимо линейной интерполяции существует множество других: тригонометрическая, квадратичная, полиномиальная, интерполяция сплайнами и другие. Какой вид интерполяции выбирать в данном случае? 

Здесь необходимо учитывать два фактора: 

  * функция $g(s,\theta)$ (часто, но не всегда) лежит в образе преобразования Радона, 
    т. е.  $g = Rf$ для некоторой функции $f$, поэтому интерполяция должна "уважать" образ $\mathrm{Im}(R)$. 
    Это значит, что интерполированные значения не должны выходить за образ $\mathrm{Im}(R)$ - такое 
    интерполирование представляет некоторые 
    сложности из-за особенностей описания $\mathrm{Im}(R)$ и в данном задании мы не будем заниматься этим вопрсом 
    (но с другой стороны, если у вас будет интерес, это можно будет обсудить)
    
  
  * Оператор $R$ - сглаживающий (более подробно мы это обсудим дальше в курсе), 
    но говоря проще - функции в образе более 
    гладкие чем их прообразы. Поэтому, в целом, для интерполяции $g(s,\theta)$ надо выбирать методы интерполяции 
    более высокого порядка чем класс гладкости в области определения. 
    Например, если мы предполагаем, что аргумент $f$ лежит в классе $C^1$, то линейной интерполяции 
    для $g(s,\theta)$ недостаточно и нам небходимо выбрать, например, квадратичную интерполяцию. 
    
В итоге, необходимо смотреть на то, что мы хотим получить в конкретной задаче. Для наших "игрушечных" примеров вполне достаточно линейной интерполяции. Также, из-за того, что мы выбираем достаточно плотную сетку, ошибка связанная с "выходом" из образа также не будет велика. Интересно, что ни один из этих пунктов мы не учитываем, но 
это следует держать в голове, что при возникновении численных проблем и появлении артефактов - интерполяция может быть одной из причин.

### 3.1 Матрица преобразования $R$
Так как мы предполагаем, что 
изображение - локально-постоянная функция с постоянными значениями внутри своих пикселей, 
то интеграл вдоль луча можно записать в виде следующей конечной суммы:

\begin{equation}
    Rf(s_i, \theta_j) = \sum\limits_{k=1}^{N_p}f_k L_k(s_i, \theta_j),
\end{equation}

где $f_k$ - значение $f$ в пикселе $k$, $L_k(s_i, \theta_j)$ - длина пересечения прямой $(s_i, \theta_j)$ с 
пикселем $k$.

<img src="./seminar1-materials/matrix-model.png" width="450" heigth="400" align="center">
<center>Преобразование Радона от изображения $f$. Здесь $f$ представлено в виде матрицы $f_{ij}$, <br> 
    но мы будем рассматривать её как одномерный линейный массив.</center>

Исходя из формулу выше, преобразование Радона можно рассматривать как умножение вектора на матрицу: 

\begin{equation}
    Rf = Lf, \text{ где } L \text{ - матрица из длин пересечений } \{L_k\}, \, L\in 
    \mathrm{Mat}_{N_sN_\theta, N_p}(R), 
\end{equation}

где $f$ - вектор размерности $N_p = N^2$ соответствующий нашему изображению.

Таким образом, $R$ является линейным оператором $R : R^{N_p} \rightarrow R^{N_sN_\theta}$. Этот оператор можно исследовать  с точки зрения линейной алгебры и численного анализа, чем мы займемся в следующей секции,

### 4. Обращение преобразований Радона при помощи матричных представлений. Стабильность восстановлений

Рассмотрим математичкский вопрос обращения преобразования Радона. Пусть нам известно преобразование $g = Rf$ и мы знаем матрицу $R$, нам необходимо восстановать $f$. Очевидно, что это задача эквивалентна решению линейного уравнения:

\begin{equation}
    Rf = g.
\end{equation}

Функция в NumPy, которая решает линейную систему это ``np.linalg.solve(A, b)``, где $A$ - матрица, $b$ - столбец данных. Если матрица $A$ - сингулярна ($\det A = 0$ с машинной точностью), тогда можно искать решение всмысле наименьших квадратов:

$$
    f^{+} = \arg \min_f \|Rf-g\|^2
$$
Такое решение всегда существует и единственно. Функция $f^{+}$ называется обобщенным решением, и известно как псевдообращение $g$ в смысле Мура-Пенроуза. Соответствующая функция в Numpy `np.linalg.lstsq(...)`.

### 5. Стабилизация при помощи SVD

В предыдущем задании мы узнали, что матрица преобразования $R$ сингулярна, а значит и задачу восстановления $f$ из $Rf$ нельзя понимать в обычном смысле. В итоге мы искали обобщенное решение $f^+$.

Дело в том, что $R^{-1}$ уже не может быть ограниченным оператором (при применении $R^{-1}$ к некоторым векторам $g$ результат $R^{-1}g$ может "взрываться"). 

**Но почему работает метод наименьших квадратов, и почему он показывает неудовлетворительные результаты при наличии шума?**

Алгоритм ``np.linalg.lstsq(...)`` восстанавливает функцию $f$ по методу наименьших квадратов. Несложно показать, что эта задача эквивалентна решению следующей системы линейных уравнений: 

$$
    R^*Rf = R^*g. 
$$

Оператор $R^*R$ - самосопряженный, а значит у него есть базис из собственных векторов, где собственные значения - это как раз сингулярные значения оператора $R$. Пусть $\{u_i\}_{i=1}^{n}, \{\sigma_i\}_{i=1}^{n}$ - ортонормированный базис из собственных векторов, тогда фунцию $f$ и $g$ можно разложить по этим базисам и воспользовавшись уравнением выше, мы получим разложение функции $f$: 

$$
    \langle f, u_i \rangle = \dfrac{\langle R^*g, u_i \rangle}{\sigma_i}.
$$

Таким образом, чем меньше $\sigma_i$, тем более нестабильно восстановление i-ой компоненты $f$. Действительно, пусть мы к $g$ добавили гауссовский шум с независимыми компонентами $\varepsilon \sim \mathcal{N}(0, \sigma^2)$. Тогда, вместо настоящей i-ой компоненты мы восстанавливаем величину :

$$
    \langle \tilde{f}, u_i \rangle := \dfrac{\langle R^*g, u_i \rangle}{\sigma_i} + \dfrac{\varepsilon}{\sigma_i}
    = f_i + \dfrac{\varepsilon}{\sigma_i}. 
$$

Из-за того, что шум $\varepsilon$ равномерно проецируется на любые собственные вектора $\{u_i\}$, для малых $\sigma_i$, член $\varepsilon / \sigma_i$ может существенно искажать наши восстановления. Специалисты в области обратных задач ещё говорят, что в таком случае шум $\varepsilon$ имеет **высокочастотную компоненту**. С другой стороны, при условии $\sigma << \sigma_i$ - восстановление i-ой компоненты сигнала стабильно.

Именно поэтому в предыдущих восстановлениях, искажения были достаточно существенны. Можно попробовать улучшить качество восстановлений, если просто не восстанавливать все компоненты: те $\sigma_i$, которые достаточно велики, можно попытаться восстановить, а те которые малы - мы просто выбросим, так как они портят наш сигнал и нет возможности их устойчиво восстановить.

$$
    f = \sum_{i=1}^{n_0} u_i \dfrac{\langle R^*g, u_i \rangle}{\sigma_i}, n < n_0.
$$

В задании 4.4 мы строили график сингулярных значений, таким образом мы можем выбрать $n_0$ эмпирически. Вопрос автоматического выбора $n_0$ - тема многочисленных исследований, и ответ зависит от многих параметров (сингулярные значения, модель шума, критерий оптимальности). В литератруе, вышеприведенная формула восстановлений с "обрезанием" до индекса $n_0$ называется регуляризацией на основе сингулярного разложения.
